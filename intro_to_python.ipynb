```python
# Create and call a function
# the function can be defined almost anywhere, as long as it is defined before it gets used

def make_strings_lowercase(str_input):

    """Form a complex number.

    Keyword arguments:
    str_input -- input string 
    """
    
    #check if input is a string
    if not isinstance(str_input, str):  
        raise TypeError('Not a string!')
    
    else:
        return str_input.lower()

#call function
print (make_strings_lowercase("UPPER CASE"))
print (make_strings_lowercase("PYTHON"))
print (make_strings_lowercase(123))
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def load_data(fname):
    
    """
    Return Preprocessed Crime Data (Pandas DataFrame)
    
    Keyword arguments:
    fname -- input data file name 
    
    """
    print('Loading Data...')
    raw = pd.read_csv('data/{}'.format(fname),sep=',',dtype={'OBJECTID':str,'TRACTCE10':str,})
    
    # Dimensions of Data
    print('Total Features: {}'.format(raw.shape[1]))
    print('Total Observations: {}'.format(raw.shape[0]))
    
    #Reform Date to DateTime Object
    print('\nReformatting Dates...')
    raw['date'] = pd.to_datetime(raw['date'])
 
    return(raw)

if __name__ =='__main__':
    
    %time df = load_data("Crime_SDCounty.csv")
```

```python
from itertools import product

def get_summary_stats(df):
    
    """
    Return Summary Descriptive Statistics for Continuous Variables
    
    Keyword arguments:
    df -- input dataframe
    """
    
    cols_interest = ['POP100','HU100','lampdist']
    
    print('==== Summary Statistics ====')
    for col,stat in product(cols_interest,['mean','var','min','max']):
        print('{0} --> {1}: {2:.3f}'.format(col,stat.capitalize(),pd.eval('df.{}.{}()'.format(col,stat))))
        
get_summary_stats(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
```

```python
import numpy as np
def population_spread(df):
    
    """
    Return Boxplot of Population Variance per City
    
    Keyword arguments:
    df -- input dataframe
    """
    
    #Gather city and population columns 
    pop_city = df[['city','POP100']].dropna().reset_index(drop=True)

    #Log Population
    pop_city['POP100'] = np.log(pop_city['POP100'])
    
    #Setup Visualization
    fig = plt.figure(figsize=(20,8))
    ax = fig.add_subplot(111)
    sns.boxplot(data=pop_city,x='city',y='POP100',ax=ax,palette='Blues')
    ax.set_ylim(0,10)
    ax.set_ylabel('Log POP100',fontsize=18)
    ax.set_xlabel('City',fontsize=18)
    plt.xticks(rotation=20)
    plt.title('Census Population Variance per City',fontsize=20,y=1.02)
    
population_spread(df)
```

**Historical Crimes from 2008 to 2013 for Specific Cities in San Diego County**

```python
def historical_crime(df,city,color):
    
    """
    Return Time Series Visualization of Daily Crime Statistics
    
    Keyword arguments:
    df -- input dataframe
    city -- input city of interest
    """
    
    d = df[df['city'] == city].groupby(['date'])['Crime_Indicator'].sum()
    
    fig = plt.figure(figsize=(20,8))
    ax = fig.add_subplot(111)
    ax.plot(d.index,d,lw=.8,marker='.',c=color,label='Reported Crime')
    
    # Rolling Median --> Window size =  2
    d.resample('M').median().rolling(window=2).median().plot(lw=4.5,c='r',ax=ax,label='Trend')
    
    ax.set_ylabel('Total Crime Count',fontsize=16)
    ax.set_xlabel('Date',fontsize=16)
    ax.legend()
    ax.set_title('City of {} Historical Daily Crime Count'.format(city),fontsize=26,y=1.02)
```

```python
#San Diego
historical_crime(df,'SndSDO','teal')
```

**San Diego County Crime History per Month**

```python
#Create Pivot Table

g = df.groupby(['city',df['date'].dt.month])['Crime_Indicator'].count().reset_index(drop=False)\
.rename(columns={'Crime_Indicator':'count'})\
.pivot(index='date',columns='city',values='count')

from sklearn.preprocessing import StandardScaler
for c in g:
    g[c] = StandardScaler().fit_transform(g[c].astype(np.float64).reshape(-1,1))
    
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)
for c in g.columns:
    ax.plot(g.index,g[c],label=c,lw=.7,alpha=0.80)
    ax.set_xlim(1,12)
    
g['Median'] = g.median(axis=1)
ax.plot(g.index,g['Median'],lw=5,c='r',ls='--',label='Median Trend')
ax.legend(loc=(1.04,0.25))
ax.set_xlabel('Month',fontsize=18)
ax.set_ylabel('Standardized Monthly Crime Rate',fontsize=18)
ax.set_title('San Diego County Crime Historical Trends per Month',fontsize=28,y=1.02)
```

**Crime Type Totals Over the 5 Year Study Period (i.e. Entire County)**

```python
from matplotlib import cm
def crime_type_totals(df):
    
    """
    Return visualization of crime totals per category for the entire county 
    
    Keyword arguments:
    df -- input dataframe
    """
    
    gp = df.groupby([df['date'].dt.month,df['date'].dt.year,'type'])['Crime_Indicator'].sum().unstack()
    
    crime_type = df['type'].dropna().unique()

    #Set up Visualization    
    n_cols = 2    
    n_rows = np.ceil(len(crime_type)/n_cols)

    fig = plt.figure(figsize=(8*n_cols,5*n_rows))
    for idx, crime in enumerate(crime_type):
        
        ax = fig.add_subplot(n_rows,n_cols,idx+1)
        colors = cm.spectral(float((idx+1)/len(crime_type)))
        
        ax.plot(np.arange(0,gp[crime].shape[0],1),gp[crime],lw=1.5,alpha=0.30,ms=10,color=colors)
        ax.plot(np.arange(0,gp[crime].shape[0],1),gp[crime].rolling(5).mean(),color=colors,lw=2.0)
        ax.set_ylabel('Time Index',fontsize=14)
        ax.set_ylabel('Reported Crime Count',fontsize=14)
        ax.set_title('Crime Type: {}'.format(crime),fontsize=16)
        
crime_type_totals(df)
```

**Total Crime Occurrences Anomaly Detection**

```python
def anomaly_crime(df,city):

    #Total Crimes Reported per Day
    d = df[df['city'] == city].groupby(['date'])['Crime_Indicator'].sum()

    #Bollinger Bands
    rolling_med = d.resample('D').median().rolling(window=30).mean()
    rolling_std = d.resample('D').median().rolling(window=30).std()

    MAD_upper = rolling_med + 1*np.std(d)
    MAD_lower = rolling_med - 1*np.std(d)

    #Provide Class Labels for Anomalies
    labels = np.where(d > MAD_upper,1,0)

    #Generate Visualization
    fig = plt.figure(figsize=(20,8))
    ax = fig.add_subplot(111)
    ax.scatter(d.index,d,c=labels,cmap='cool')
    ax.plot(d.index,rolling_med,lw=2.5,c='teal',label='Rolling Median')
    ax.plot(d.index,MAD_lower,lw=2.5,c='red',label='Upper Band')
    ax.plot(d.index,MAD_upper,lw=2.5,c='red',label='Lower Band')
    ax.set_ylabel('Daily Crime Count',fontsize=16)
    ax.set_xlabel('Date',fontsize=16)
    ax.legend()
    ax.set_title('City of {} Anomaly Crime History'.format(city),fontsize=26,y=1.02)
    
anomaly_crime(df,'SndSDO')
```

```python
import numpy as np
import numpy.linalg as lin

#Create Matrix --> 3x3
a = np.array([[0,1,6],
              [3,4,8],
              [5,6,1]])

print('\nOriginal Matrix:')
print(a)
print('\nTranspose Matrix:')
print(a.T)
print('\nInverse Matrix:')
print(np.invert(a))
print('\nFlatten Matrix:')
print(a.flatten())
print('\nTrace of Matrix:')
print(np.trace(a))
print('\nDiag of Matrix:')
print(np.diag(a))
print('\nDeterminent of Matrix:')
print(lin.det(a))
```

```python
print('Matrix A: \n',a)
val,vect = lin.eig(a)
print('\nEigenvalues of A:\n ',val)
print('\nEigenvector of A:\n', vect)
```
